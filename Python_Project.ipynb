```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Define Countries

```python
countries = ["United States", "India", "Brazil", "Germany", "South Africa"]
```

```python
df = df[df["location"].isin(countries)]
```

Data Cleaning

Drop duplicate rows

```python
 df.drop_duplicates()
```

Drop rows with all missing values

```python
df.dropna(how='all')
```

Fill missing values in specific columns (customize as needed)

```python
df.fillna({
        'cases': 0,
        'deaths': 0,
        'recovered': 0
    })
```

 Strip leading/trailing whitespace from column names

```python
df.columns.str.strip()
```

Convert date columns to datetime

```python
if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

Remove rows with invalid dates if any

```python
 df = df[df['date'].notna()] if 'date' in df.columns else df
```

Filter countries of Interest(Kenya, USA, India)



```python
#Data Cleaning Part
intrestCountries=['Kenya','USA','INDIA']
df = df[df['location'].isin(intrestCountries)]
df.dropna( inplace=True)
df['date'] = pd.to_datetime(df['date'])
df_filtered = df[df['location'].isin(intrestCountries)]
df_filtered.dropna(inplace=True)
# Handle missing numeric values
numeric_columns = df.select_dtypes(include=['number']).columns.tolist()
df[numeric_columns] = df[numeric_columns].fillna(method='ffill')

print("Missing values in each column after cleaning:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df[df["location"] == country]
    plt.plot(country_data["date"], country_data["total_cases"], label=country)
plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
latest_data = df[df["date"] == df["date"].max()]
top_countries = latest_data.groupby("location")["total_cases"].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Reds_r")
plt.title("Top 10 Countries by Total COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
correlation_data = df[df["location"].isin(countries)][["new_cases", "new_deaths", "total_cases", "total_deaths"]].dropna()
corr_matrix = correlation_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of COVID-19 Metrics")
plt.tight_layout()
plt.show()
```

```python
# Get latest data for each country
latest_vax = df.sort_values("date").groupby("location").tail(1)

# Use people_vaccinated_per_hundred for % comparison
plt.figure(figsize=(10, 6))
sns.barplot(data=latest_vax, x="people_vaccinated_per_hundred", y="location", palette="viridis")
plt.xlabel("% of Population Vaccinated")
plt.title("Vaccinated Population Percentage by Country")
plt.tight_layout()
plt.show()
```

```python
fig = px.line(
    df,
    x="date",
    y="total_vaccinations",
    color="location",
    title="Cumulative COVID-19 Vaccinations Over Time",
    labels={"total_vaccinations": "Total Vaccinations", "location": "Country"}
)
fig.show()
```

```python
latest = df.sort_values("date").groupby("location").tail(1)

fig = px.bar(
    latest,
    x="location",
    y="people_vaccinated_per_hundred",
    color="location",
    title="% of Population Vaccinated (Latest)",
    labels={"people_vaccinated_per_hundred": "% Vaccinated"}
)
fig.show()
```
